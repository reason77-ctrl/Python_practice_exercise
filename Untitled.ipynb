```python
from bs4 import BeautifulSoup
url = BeautifulSoup('http://www.nepalstock.com/todaysprice', 'html.parser')
url
```

```python
import requests
response = requests.get(url)
response
```

```python
soup = response.text
soup = BeautifulSoup(soup,"lxml")
soup
```

```python
soup.table
```

```python
table = soup.table
```

```python
tags = table.find_all('tr')
tags
```

```python
data = []
for i in tags:
    data.append(i.text)
    print(data)
```

```python
data = data[1:-3]
```

```python
data
```

```python
for i in data:
    print(i.split('\n')[1:-1])
```

```python
final_output = []
for i in data:
    output = i.split('\n')[1:-1]
    final_output.append(output)
```

```python
final_output
```

```python
import csv
with open('stock.csv','a') as f:
    x = csv.writer(f)
    for i in final_output:
        x.writerow(i)
```
